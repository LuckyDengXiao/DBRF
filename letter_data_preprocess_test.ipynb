```python
with open(data_path) as f:
    rows = [row.strip().split(',') for row in f.readlines()]
```

【osp.join】将文件夹路径连接起来，可以使用abspath()来调取绝对路径

```python
print(rows)
```

```python
n_datas = len(rows)
X = np.zeros((n_datas, 16), dtype=np.float32)
y = np.zeros(n_datas, dtype=np.int32)
```

```python
for i, row in enumerate(rows):
    X[i, :] = list(map(float, row[1:]))
    y[i] = ord(row[0]) - ord('A')
```

【row[1:]】row是一行，一维数组，1：代表从下标1开始取到结尾。

【ord()】 函数是 chr() 函数（对于8位的ASCII字符串）或 unichr() 函数（对于Unicode对象）的配对函数，它以一个字符（长度为1的字符串）作为参数，返回对应的 ASCII 数值

【X[i,:]】X是二维数组，i,:代表下标i行，从0开始取到结尾。
